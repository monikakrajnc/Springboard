```python
# import libraries

import pandas as pd
import json 
from pandas.io.json import json_normalize
```

```python
# import data

original_json_df = pd.read_json('data/world_bank_projects.json')
```

```python
#original_json_df.info()
```

```python
#original_json_df.columns
```

```python
themes_column = original_json_df['mjtheme_namecode']
```

```python
#themes_column
```

```python
# create a list with all projects

all_projects = []
for ent in themes_column:
    for dictionary in ent:
        all_projects.append(dictionary)
```

```python
# use normalization to create a table from nested elements

normalized_data = json_normalize(all_projects)
```

```python
#normalized_data
```

```python
# create a list with projects that have name for the project theme (with the poject theme name)

with_name = []

for index, row in normalized_data.iterrows():
    if not row['name'] == '':
        with_name.append(row)

        
# create a list with all distinct project themes and sort that list

all_unique_project = list({v['code']:v for v in with_name}.values())
sorted_all_unique_project = sorted(all_unique_project, key=lambda k: k['code']) 
```

```python
# go through each row in the table and add name for a project theme, where is missing

for index, row in normalized_data.iterrows():
    if row['name']== '':
        for uni in sorted_all_unique_project:
            if row['code'] == uni['code']:
                row['name'] = uni['name']
```

```python
# dataframe with filled in missing values for project theme names

normalized_data
```
