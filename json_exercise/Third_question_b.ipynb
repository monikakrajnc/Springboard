```python
# import libraries

import pandas as pd
import json 
from pandas.io.json import json_normalize
```

```python
# import data

original_json_df = pd.read_json('data/world_bank_projects.json')
```

```python
# select 'mjtheme_namecode' column from the dataframe

project_themes = original_json_df['mjtheme_namecode']
```

```python
# create a list of all projects

all_projects = []
for entry in project_themes:
    for dictionary in entry:
        all_projects.append(dictionary)
```

```python
# create a list with projects that have a project theme name

with_name = []

for elem in all_projects:
    if not elem['name'] == '':
        with_name.append(elem)

        
# create a list with all distinct project themes and sort that list

all_unique_project = list({v['code']:v for v in with_name}.values())
sorted_all_unique_project = sorted(all_unique_project, key=lambda k: k['code']) 
```

```python
# check each project if it has a name
# if not, check its code and then find the code in 'sorted_all_unique_project' 
# and select the name for the project theme that represents the project code

# go through each element in project_themes
for element in project_themes:
    for dictionary in element:
        # if dictionary doesn't include the name for the project theme (it has a missing value)...
        if dictionary['name'] == '':
            # ...go through each element in sorted_all_unique_project
            for uni in sorted_all_unique_project:
                # find element 'uni' that has the same code as the one (in dictionary) without the name for the project theme
                if dictionary['code'] == uni['code']:
                    # assign the name from 'uni' element to the 'dictionary' element
                    dictionary['name'] = uni['name']
```
